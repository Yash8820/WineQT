# Let's find outliers

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Attributes_scaled = pd.DataFrame(scaler.fit_transform(Attributes),columns=Attributes.columns)
Attributes_scaled.boxplot(return_type = 'axes',figsize=(20,20),vert=False)
```

## We can see that alcohol and citric acids (fields) are well maintained

```python
Attributes_scaled.drop(Attributes_scaled[['Id']],axis=1,inplace=True)
```

# Summary:
## There is no missing values in the data
## Max number of wines achieved the quality scale of 5 & 6
## There are some correlation between features
## More features consists of outliers

```python
Attributes_scaled
```

# Splitting our data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Attributes_scaled,classification,test_size=0.3,random_state=4)
```

```python
len(x_train)
```

```python
len(x_test)
```

# Applying different model

```python
from sklearn import tree
```

```python
from sklearn import linear_model
```

## Logistic regression

```python
model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

## Accuracy is 59%

## Decision tree

```python
model2=tree.DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

## Accuracy is 54%

## Random forest

```python
from sklearn import ensemble
```

```python
model4=ensemble.RandomForestClassifier(n_estimators=30)
model4.fit(x_train,y_train)
```

```python
model4.score(x_test,y_test)
```

## Accuracy is 64%

## SVM

```python
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
```

```python
model5 = svm.SVC(C=1,gamma="auto",kernel="linear")
model5.fit(x_train,y_train)
```

```python
model5.score(x_test,y_test)
```

## Accuracy is 57%

## KNN

```python
model6 = KNeighborsClassifier(n_neighbors=7)
model6.fit(x_train,y_train)
```

```python
model6.score(x_test,y_test)
```

## Accuracy is 57%
